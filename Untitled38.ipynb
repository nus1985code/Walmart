```python
# Columns you want to process
selected_columns = ['Weekly_Sales', 'Unemployment', 'Temperature']

# Calculate Q1, Q3, and IQR
Q1 = df[selected_columns].quantile(0.25)
Q3 = df[selected_columns].quantile(0.75)
IQR = Q3 - Q1

# Calculate lower and upper limits
Lower_limit = Q1 - 1.5 * IQR
Upper_limit = Q3 + 1.5 * IQR

# Define the limit_imputer function
def limit_imputer(value, lower_limit, upper_limit):
    return min(max(value, lower_limit), upper_limit)
# Apply the limit_imputer function element-wise to the selected columns
for column in selected_columns:
    df[column] = df[column].apply(limit_imputer, lower_limit=Lower_limit[column], upper_limit=Upper_limit[column])

# Print the processed DataFrame
print(df)
```

```python
#Create Boxplot
for i in df.columns:
  if (df[i].dtypes == 'int64' or df[i].dtypes == 'float64'):
    sns.boxplot(df[i])
    plt.show()
```

```python
# seperate Date into month ,week and year and add it to the dataframe df
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df['Month'] = df['Date'].dt.month
df['Week'] = df['Date'].dt.isocalendar().week
df['Year'] = df['Date'].dt.year
```

```python
# which weekly sales are affected by the unemployment rate, if yes - which stores are suffering the most? show in infgraphic and also in percentage if possible df['Week'] = df['Date'].dt.isocalendar().week

# Calculate the correlation between Weekly_Sales and Unemployment
correlation = df['Weekly_Sales'].corr(df['Unemployment'])

# Print the correlation
print(f"Correlation between Weekly_Sales and Unemployment: {correlation}")

# Identify stores with the lowest Weekly_Sales
lowest_sales_stores = df.groupby('Store')['Weekly_Sales'].mean().sort_values(ascending=True).head(5)

# Create a bar chart of average Weekly_Sales by store
plt.figure(figsize=(12, 6))
sns.barplot(x=lowest_sales_stores.index, y=lowest_sales_stores.values)
plt.xlabel('Store')
plt.ylabel('Average Weekly Sales')
plt.title('Stores with the Lowest Average Weekly Sales')
plt.show()

# Calculate the percentage change in Weekly_Sales for each store when Unemployment increases by 1%
sales_change_by_unemployment = df.groupby('Store')['Weekly_Sales'].apply(lambda x: (x.pct_change() / df['Unemployment'].pct_change()).mean() * 100)

# Print the percentage change for each store
print(f"Percentage change in Weekly_Sales for each store when Unemployment increases by 1%:")
print(sales_change_by_unemployment)

```

```python
# If the weekly sales show a seasonal trend, when and what could be the reason? df['Week'] = df['Date'].dt.isocalendar().week

# Calculate the average weekly sales for each week of the year
average_sales_by_week = df.groupby('Week')['Weekly_Sales'].mean()

# Create a line chart of average weekly sales by week
plt.figure(figsize=(12, 6))
sns.lineplot(x=average_sales_by_week.index, y=average_sales_by_week.values)
plt.xlabel('Week')
plt.ylabel('Average Weekly Sales')
plt.title('Average Weekly Sales by Week')
plt.show()

```

```python
# Does temperature affect the weekly sales in any manner?

# Calculate the correlation between Weekly_Sales and Temperature
correlation = df['Weekly_Sales'].corr(df['Temperature'])

# Print the correlation
print(f"Correlation between Weekly_Sales and Temperature: {correlation}")

# Calculate the percentage change in Weekly_Sales for each store when Temperature increases by 1 degree
sales_change_by_temperature = df.groupby('Store')['Weekly_Sales'].apply(lambda x: (x.pct_change() / df['Temperature'].pct_change()).mean() * 100)

# Print the percentage change for each store
print(f"Percentage change in Weekly_Sales for each store when Temperature increases by 1 degree:")
print(sales_change_by_temperature)

# Create a scatter plot of Weekly_Sales vs Temperature
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df['Temperature'], y=df['Weekly_Sales'])
plt.xlabel('Temperature')
plt.ylabel('Weekly Sales')
plt.title('Scatter Plot of Weekly Sales vs Temperature')
plt.show()
```

The scatter plot indicates that weekly sales at Walmart do not have a strong direct correlation with temperature. While sales can be high or low across the entire temperature range, there are dense clusters of moderate sales figures. This suggests that other factors, such as marketing campaigns, regional differences, and product seasonality, might play more significant roles in influencing weekly sales than temperature alone.








```python
#  How is the Consumer Price index affecting the weekly sales of various stores?

# Calculate the correlation between Weekly_Sales and Consumer Price Index (CPI)
correlation = df['Weekly_Sales'].corr(df['CPI'])

# Print the correlation
print(f"Correlation between Weekly_Sales and CPI: {correlation}")

# Calculate the percentage change in Weekly_Sales for each store when CPI increases by 1%
sales_change_by_cpi = df.groupby('Store')['Weekly_Sales'].apply(lambda x: (x.pct_change() / df['CPI'].pct_change()).mean() * 100)

# Print the percentage change for each store
print(f"Percentage change in Weekly_Sales for each store when CPI increases by 1%:")
print(sales_change_by_cpi)

# Create a scatter plot of Weekly_Sales vs CPI
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df['CPI'], y=df['Weekly_Sales'])
plt.xlabel('CPI')
plt.ylabel('Weekly Sales')
plt.title('Scatter Plot of Weekly Sales vs CPI')
plt.show()
```

The scatter plot indicates that weekly sales at Walmart do not have a straightforward correlation with CPI, but rather, they are influenced by a range of factors. The existence of two distinct CPI clusters suggests that there may be underlying conditions or periods that separate the data. Sales are spread across a wide range within each cluster, pointing to the influence of various factors beyond CPI on weekly sales.

```python
#Transformation(log)first_log
Sales_log = np.log(Sales['Weekly_Sales'])
Sales_log =  Sales_log.dropna()
Sales_log
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Assuming df is your main dataframe containing 'Store', 'Date', and 'Weekly_Sales'

# Filter the data for Store 20
store_20_data = df[df['Store'] == 20].set_index('Date')

# Ensure the data is sorted by Date and there are no missing values
store_20_data = store_20_data.sort_index()
store_20_data = store_20_data.asfreq('W', method='pad')  # Ensure weekly frequency

# Check for any remaining missing values
if store_20_data.isnull().sum().sum() > 0:
    print("There are missing values in the data. Please handle them before proceeding.")
else:
    # Perform train-test split
    train_size = int(len(store_20_data) * 0.8)
    train, test = store_20_data[:train_size], store_20_data[train_size:]

    # Fit the SARIMAX model with seasonal order for 12 weeks
    model = SARIMAX(train['Weekly_Sales'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
    model_fit = model.fit(disp=False)

    # Forecast for the next 12 weeks
    forecast_steps = 12
    forecast = model_fit.get_forecast(steps=forecast_steps)
    forecast_index = pd.date_range(start=train.index[-1] + pd.Timedelta(weeks=1), periods=forecast_steps, freq='W')
    forecast_series = pd.Series(forecast.predicted_mean.values, index=forecast_index)

    # Combine the train, test, and forecast for visualization
    combined = pd.concat([train['Weekly_Sales'], test['Weekly_Sales'], forecast_series], keys=['Train', 'Test', 'Forecast'])

    # Plot the results
    plt.figure(figsize=(12, 6))
    plt.plot(train.index, train['Weekly_Sales'], label='Train')
    plt.plot(test.index, test['Weekly_Sales'], label='Test')
    plt.plot(forecast_series.index, forecast_series, label='Forecast', color='green')
    plt.xlabel('Date')
    plt.ylabel('Weekly Sales')
    plt.title('Train, Test, and Forecasted Weekly Sales for Store 20')
    plt.legend(loc='best')
    plt.show()

    # Check the contents of forecast_series to ensure it's not empty
    print(forecast_series)
```
